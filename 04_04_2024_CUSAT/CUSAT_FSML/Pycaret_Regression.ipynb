```python
from pycaret.datasets import get_data
from pycaret.regression import setup, compare_models, save_model, load_model, predict_model
import pandas as pd

# Load the insurance dataset
data = get_data('insurance')

# Set up the regression experiment
s = setup(data, target='charges', session_id=123)

# Compare and select the best model
best_model = compare_models()

# Save the best model
save_model(best_model, 'best_insurance_model')
```

```python
# Load the saved model
loaded_model = load_model('best_insurance_model')
```

```python
# Sample new data for prediction (example)
new_data = pd.DataFrame({
    'age': [45, 55],
    'sex': ['male', 'female'],
    'bmi': [30, 25],
    'children': [1, 2],
    'smoker': ['yes', 'no'],
    'region': ['southeast', 'southwest']
})

# Make predictions on new data using the loaded model
predictions = predict_model(loaded_model, data=new_data)

# Display predictions
print(predictions)
```
